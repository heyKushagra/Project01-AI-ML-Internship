# **`Cleaning the Data`**



```python
df['director'] = df['director'].fillna('Unknown')
df['cast']=df['cast'].fillna('Unkown')
df['country']=df['country'].fillna('Unknown')
```

```python
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
```

```python
df.duplicated().sum()
```

> Replaced missing values with "Uknown" in Director, Cast and Country Column. And removed missing values row from date_added, rating and duration column.

```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    format='mixed'
)
```

```python
print(df.shape)
df.isnull().sum()
```

# **`Exploratory Data Analysis`**



```python
df['type'].value_counts()
```

```python

plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)

plt.title('Distribution of Movies and TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')

plt.show()
```

Netflix contains significantly more Movies than TV Shows, indicating that movies make up the majority of the platform's content library.

```python
top_countries = df['country'].value_counts().head(10)
top_countries
```

```python
top_countries.sort_values().plot(kind='barh')
plt.title("Top 10 Countries by Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

The United States is expected to dominate the content library, followed by a few major content-producing countries.

```python
df['year_added'] = df['date_added'].dt.year
```

```python
content_by_year = df['year_added'].value_counts().sort_index()
content_by_year
```

```python
plt.figure(figsize=(12,6))
plt.plot(content_by_year.index,
         content_by_year.values, marker='o')

plt.title('Netflix Content Added Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)

plt.show()
```

Netflix experienced rapid content growth during the late 2010s, indicating aggressive expansion of its content library.

```python
top_ratings = df['rating'].value_counts()
top_ratings
```

```python
plt.figure(figsize=(8,8))
top_ratings.head(8).plot(kind='pie', autopct='%1.1f%%')
plt.title (' Distribution of Top Netflix Content Ratings')
plt.ylabel('')
plt.show()
```

TV-MA and TV-14 generally dominate the platform, suggesting Netflix targets mature and teenage audiences.

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()

top_genres = genres.value_counts().head(10)

top_genres
```

```python
top_genres.sort_values().plot( kind = 'barh')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

Drama, International Movies, and Comedy are among the most frequently occurring genres on Netflix.

```python
plt.figure(figsize=(10,6))

plt.hist(
    df['release_year'],
    bins=30
)

plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.show()
```

Most Netflix content is concentrated in recent years, with significantly fewer titles from older decades.

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    df[['release_year','year_added']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

There is generally a positive relationship between release year and the year content was added to Netflix, indicating newer content tends to be added sooner.

# **Business Insights**

> 1. Movies dominate Netflix's catalog, suggesting the platform focuses heavily on film content to maximize catalog size and viewer choice.

> 2. Content additions increased sharply during the late 2010s, indicating an aggressive expansion phase as Netflix competed globally in the streaming market.

> 3. A small number of countries contribute most content, showing that Netflix's library is concentrated around major entertainment-producing regions.

> 4. TV-MA and TV-14 ratings account for a large share of titles, suggesting Netflix primarily targets teenage and adult audiences rather than young children.

> 5. Drama, International Movies, and Comedy are among the most common genres, indicating that Netflix emphasizes broadly appealing content with global audience potential.
